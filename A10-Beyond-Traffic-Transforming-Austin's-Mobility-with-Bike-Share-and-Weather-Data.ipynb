## **Weather-Specific Trends**

**How does the likelihood of bike trips change on days with varying ranges of humidity?**

The SQL query below investigates how bike usage varies with humidity levels by joining bike trip data with corresponding daily weather data. It classifies each trip day into three humidity categories: 'Low Humidity' (humidity < 50%), 'Moderate Humidity' (humidity between 50% and 70%), and 'High Humidity' (humidity > 70%). For each humidity level, the query calculates the total number of bike trips, the average trip duration in minutes, and the average number of trips per day by dividing the trip count by the number of unique days in each humidity category. By organizing and comparing these metrics across humidity levels, this code provides insights into whether and how bike ridership patterns (frequency and trip duration) change with different humidity conditions.

```python
%%bigquery
WITH trips_weather AS (
  SELECT
    trips.trip_id,
    DATE(trips.start_time) AS trip_date,
    trips.start_station_name,
    trips.end_station_name,
    trips.duration_minutes,
    trips.subscriber_type,
    weather.humidity
  FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS trips
  JOIN
    `fall24-ba775-a10.ba_775_a10_project.austin_weather` AS weather
  ON
    DATE(trips.start_time) = DATE(weather.datetime)
)

SELECT
  CASE
    WHEN humidity < 50 THEN 'Low Humidity'
    WHEN humidity BETWEEN 50 AND 70 THEN 'Moderate Humidity'
    ELSE 'High Humidity'
  END AS humidity_level,
  COUNT(trip_id) AS total_trips,
  ROUND(AVG(duration_minutes), 2) AS avg_duration_minutes,
  ROUND(COUNT(trip_id) / COUNT(DISTINCT trip_date), 2) AS avg_trips_per_day
FROM
  trips_weather
GROUP BY
  humidity_level
ORDER BY
  humidity_level;
```

The analysis reveals that bike ridership is generally consistent across different humidity levels, with only slight variations. On high-humidity days, the total trip count is substantial at 894,680 trips, averaging about 580 trips per day, with an average trip duration of 28.32 minutes. Similarly, on moderate-humidity days, the ridership is the highest at 1,068,255 trips, with an average of 585 trips per day and a slightly longer average trip duration of 29.39 minutes. Low-humidity days have fewer total trips at 308,217, but interestingly, the daily average trip count is highest at 639.45, though the average trip duration is the shortest at 27.54 minutes. These results suggest that while humidity does not drastically impact ridership volume, it may slightly influence trip duration, with lower humidity potentially encouraging shorter trips but with a slightly higher trip frequency on average per day. Overall, ridership remains fairly resilient to humidity variations, but small shifts in trip length and daily frequency may be observed across different humidity levels.

**What weather conditions most affect weekend vs. weekday usage patterns?**

This query analyzes how weather conditions affect bike share usage by comparing patterns on weekends and weekdays. It starts by classifying each bike trip as either "Weekend" or "Weekday" using the `DAYOFWEEK` function, which identifies trips that took place on Saturdays or Sundays as "Weekend" and all others as "Weekday." The query then calculates several aggregate statistics, including the total number of trips and average values for key weather conditions—temperature, humidity, precipitation, wind speed, and cloud cover—separately for each day type. These statistics provide insights into the relationship between bike usage and weather, helping to identify if certain weather conditions influence bike share usage differently on weekends versus weekdays. The results are ordered by day type, presenting a clear comparison of weekend and weekday patterns.

```python
%%bigquery
SELECT
  CASE
    WHEN EXTRACT(DAYOFWEEK FROM trips.start_time) IN (1, 7) THEN 'Weekend'
    ELSE 'Weekday'
  END AS day_type,
  COUNT(trips.trip_id) AS total_trips,
  AVG(weather.temp) AS avg_temp,
  AVG(weather.humidity) AS avg_humidity,
  AVG(weather.precip) AS avg_precipitation,
  AVG(weather.windspeed) AS avg_windspeed,
  AVG(weather.cloudcover) AS avg_cloudcover
FROM
  `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS trips
JOIN
  `fall24-ba775-a10.ba_775_a10_project.austin_weather` AS weather
ON
  DATE(trips.start_time) = DATE(weather.datetime)
GROUP BY
  day_type
ORDER BY
  day_type;
```

The analysis of bike share usage patterns reveals some subtle differences between weekends and weekdays in relation to weather conditions. On weekdays, there is a slightly higher average temperature (21.80°C) compared to weekends (21.63°C), but the difference is minimal. Humidity is slightly lower on weekdays (65.34%) than on weekends (65.79%). Precipitation levels are slightly higher on weekdays (1.81 mm) compared to weekends (1.49 mm), which may suggest that bike share usage on weekdays is somewhat more resilient to rain. Wind speeds and cloud cover are very similar between the two periods, with weekday wind speeds averaging 18.41 km/h and weekend wind speeds averaging 18.27 km/h, while cloud cover is nearly identical, with weekdays at 43.49% and weekends at 42.90%. Overall, weather conditions such as temperature, humidity, and wind speed do not show significant differences between weekends and weekdays, although weekday usage seems to be marginally more affected by higher precipitation. These insights suggest that weather conditions may have a somewhat similar effect on bike share usage across both weekdays and weekends, with minor variations in precipitation influencing weekday usage more.

**Which weather conditions (e.g., windy, humid) lead to the highest decrease in ridership?**

This code analyzes the impact of specific weather conditions on bikeshare ridership to identify which conditions correlate with decreased usage. The query first aggregates trips by station and day, calculating average humidity, wind gust, and cloud cover for each trip date. Then, in the `weather_ridership_analysis` CTE, the weather conditions are classified into categories: *High, Moderate,* or *Low* levels for each metric (humidity, wind gust, and cloud cover). By grouping these categories, it calculates the average trip count across different combinations of weather conditions. Finally, the main `SELECT` statement retrieves and orders the results by `avg_trip_count` in ascending order, allowing us to easily identify which weather combinations (e.g., *High Humidity* and *High Wind Gust*) have the lowest ridership. This analysis helps determine which weather conditions most significantly decrease bikeshare usage.What are the top 10 subscription plans and how are they distributed

```python
%%bigquery
WITH weather_conditions AS (
  SELECT
    trips.start_station_id,
    trips.start_station_name,
    DATE(trips.start_time) AS trip_date,
    COUNT(trips.trip_id) AS trip_count,
    AVG(weather.humidity) AS avg_humidity,
    AVG(weather.windgust) AS avg_windgust,
    AVG(weather.cloudcover) AS avg_cloudcover
  FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS trips
  JOIN
    `fall24-ba775-a10.ba_775_a10_project.austin_weather` AS weather
  ON
    DATE(trips.start_time) = DATE(weather.datetime)
  GROUP BY
    trips.start_station_id, trips.start_station_name, trip_date
),

weather_ridership_analysis AS (
  SELECT
    CASE
      WHEN avg_humidity > 80 THEN 'High Humidity'
      WHEN avg_humidity BETWEEN 60 AND 80 THEN 'Moderate Humidity'
      ELSE 'Low Humidity'
    END AS humidity_level,
    CASE
      WHEN avg_windgust > 20 THEN 'High Wind Gust'
      WHEN avg_windgust BETWEEN 10 AND 20 THEN 'Moderate Wind Gust'
      ELSE 'Low Wind Gust'
    END AS windgust_level,
    CASE
      WHEN avg_cloudcover > 70 THEN 'High Cloud Cover'
      WHEN avg_cloudcover BETWEEN 40 AND 70 THEN 'Moderate Cloud Cover'
      ELSE 'Low Cloud Cover'
    END AS cloudcover_level,
    AVG(trip_count) AS avg_trip_count
  FROM
    weather_conditions
  GROUP BY
    humidity_level, windgust_level, cloudcover_level
)

SELECT
  humidity_level,
  windgust_level,
  cloudcover_level,
  avg_trip_count
FROM
  weather_ridership_analysis
ORDER BY
  avg_trip_count ASC
```

This query investigates how precipitation affects bike usage across different stations to determine if rainy conditions lead to a decrease in ridership at each station or if certain stations maintain consistent usage regardless of weather. The code starts by aggregating bike trip data at each station and joins it with weather data to classify each day as either "rainy" or "clear" based on precipitation values. It calculates the total number of trips per station, the number of rainy and clear days, and the average number of trips taken on rainy days (`avg_rainy_day_trips`) and clear days (`avg_clear_day_trips`). Finally, it computes a metric called `precipitation_impact` by comparing the average trips on rainy days to those on clear days. If the ratio of `avg_rainy_day_trips` to `avg_clear_day_trips` is greater than 0.5, the station is labeled as having "Consistent Usage" despite rain; otherwise, it’s labeled "Reduced Usage." This categorization allows an assessment of which stations maintain stable ridership across varying weather conditions and identifies any stations less affected by precipitation.

**Are users more likely to take longer or shorter trips on cloudy days compared to clear days?**

This SQL query examines the relationship between weather conditions and average bike trip durations. It creates two datasets: **WeatherCleaned**, which transforms the `datetime` column into `weather_date` and classifies days as "Sunny" (when `preciptype` is NULL) or "Rainy" (when it is not NULL), and **BikeShareCleaned**, which extracts the date from the `start_time` column (`trip_date`) and retains trips with valid `start_time` and `duration_minutes` values. The datasets are joined on their respective date columns to associate weather conditions with bike trips. Finally, the query calculates the average trip duration (`avg_trip_duration`) grouped by weather type, enabling a comparison of bike usage patterns on sunny and rainy days.

```python
%%bigquery
WITH WeatherCleaned AS (
  SELECT
    DATE(datetime) AS weather_date,
    CASE
      WHEN preciptype IS NULL THEN 'Sunny'
      ELSE 'Rainy'
    END AS day_type
  FROM `fall24-ba775-a10.ba_775_a10_project.austin_weather`
),
BikeShareCleaned AS (
  SELECT
    DATE(start_time) AS trip_date,
    duration_minutes
  FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
  WHERE start_time IS NOT NULL AND duration_minutes IS NOT NULL
)
-- Join the tables on weather_date and trip_date
SELECT
  day_type,
  AVG(duration_minutes) AS avg_trip_duration
FROM WeatherCleaned
JOIN BikeShareCleaned
ON WeatherCleaned.weather_date = BikeShareCleaned.trip_date
GROUP BY day_type;
```

The query results indicate that users take slightly longer trips on sunny days compared to rainy days, with an average trip duration of 29.25 minutes on sunny days and 27.69 minutes on rainy days. This suggests that adverse weather conditions, such as rain, may discourage users from taking longer bike trips, likely due to reduced comfort or safety concerns. While the difference is relatively small, the pattern aligns with the expectation that clear weather encourages more leisurely and extended bike usage compared to rainy conditions.

## **Subscriber Type & Behavior Patterns**

**What are the top 5 subscription plans and how are they distributed?**

This code identifies the top 10 subscription plans and lists their respective usage counts.
First, we extract the `subscriber_type` from the `austin_bikeshare.bikeshare_trips` table and aggregate the ride count for each `subscription type`. Since the `subscriber_type` field lacks standardization, we use a CASE statement to rename the categories and then re-aggregate them to ensure data consistency.

```python
%%bigquery
WITH agg_subscribe AS (
SELECT
    CASE
        WHEN subscriber_type LIKE '%Student Membership%' THEN 'Student Membership'
        WHEN subscriber_type IN ('Local31', 'Local30') THEN 'Month Pass'
        WHEN subscriber_type LIKE 'Local365%' THEN 'Annual Pass'
        WHEN subscriber_type LIKE '%Walk Up%' THEN '1 Day Pass'
        WHEN subscriber_type IN ('Pay-as-you-ride', 'Explorer', '%Single Trip%') THEN 'Single Trip'
        WHEN subscriber_type LIKE '3-Day%' THEN '3 Day Pass'
        WHEN subscriber_type LIKE '7-Day%' THEN '7 Day Pass'
        WHEN subscriber_type = 'Weekender' THEN '7 Day Pass'
        WHEN subscriber_type LIKE '%Single Trip%' THEN 'Single Trip'
        ELSE subscriber_type
    END AS subscriber_type,
    count_subscribe_type
FROM (
      SELECT DISTINCT subscriber_type, count(*) as count_subscribe_type
      FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
      GROUP BY subscriber_type
      HAVING subscriber_type IS NOT NULL
      ORDER BY count_subscribe_type desc

      )
ORDER BY count_subscribe_type DESC

)
Select subscriber_type, SUM( count_subscribe_type) as total_subscribe_number
From agg_subscribe
GROUP BY subscriber_type
ORDER BY total_subscribe_number DESC
LIMIT 5
```

The top five subscription plans show a clear preference for short-term options, with the *Student Membership* leading significantly at 719,464 subscribers, followed by the *1 Day Pass* with 532,904 subscribers. These two short-term plans indicate a strong demand for flexible and occasional usage, likely among students and casual riders. **Which stations are the busiest during extreme weather events and may need better bike stocking?**

To identify the busiest bikeshare station during extreme weather conditions, we first categorize dates based on specific weather factors using the `CASE` statement and create a table named `ExtremeWeather` that includes only the dates that are not treated as “Normal”. (NOTE: If the maximum temperature exceeds 35 degrees or minimum temperature falls below 0 degrees, the weather is categorized as `Extreme Temperature`. If the humidity is greater than 90%, then it is categorized as `High Humidity`. If the `Rain` or `Snow` are included in the conditions column, then they are categorized as `Extreme Precipitation`. All other conditions are classified as `Normal`.) We then link trips to their starting stations by matching `station_id` between `bikeshare_stations` and `bikeshare_trips` tables. Next, we joined with `ExtremeWeather` by matching trip dates to days with extreme weather. To calculate the number of trips from each station during these events, we group by station and weather condition. Finally, the results are ordered by trip count, and we limit the output to show only the station with the highest number of trips during extreme weather conditions.

```python
%%bigquery
WITH ExtremeWeather AS (
 SELECT
   datetime,
   CASE
     WHEN tempmax > 35 OR tempmin < 0 THEN 'Extreme Temperature'
     WHEN humidity > 90 THEN 'High Humidity'
     WHEN conditions LIKE '%Rain%' OR conditions LIKE '%Snow%' THEN 'Extreme Precipitation'
     ELSE 'Normal'
   END AS weather_condition
 FROM `fall24-ba775-a10.ba_775_a10_project.austin_weather`
 WHERE
   tempmax > 35 OR tempmin < 0 OR humidity > 90 OR conditions LIKE '%Rain%' OR conditions LIKE '%Snow%'
)


SELECT
 s.station_id,
 s.name AS station_name,
 COUNT(t.trip_id) AS trip_count,
 e.weather_condition
FROM
 `bigquery-public-data.austin_bikeshare.bikeshare_stations` AS s
INNER JOIN
 `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS t
ON
 s.station_id = t.start_station_id
INNER JOIN
 ExtremeWeather AS e
ON
 DATE(t.start_time) = e.datetime
GROUP BY
 s.station_id, s.name, e.weather_condition
ORDER BY
 trip_count DESC
    LIMIT 20;
```

The data indicates that the busiest stations during extreme weather events, both precipitation and temperature, are primarily located around high-traffic areas, with "21st & Speedway @PCL" leading with a total of 44,608 trips in extreme precipitation and 30,226 trips in extreme temperature. Other highly used stations include "4th/Sabine," "Nueces & 26th," and "Guadalupe & 21st," each logging over 10,000 trips across both types of extreme conditions. Stations like "Dean Keeton & Whitis" and "UT West Mall @ Guadalupe" also show significant usage, indicating high demand even during extreme weather. These stations, located around the University of Texas campus and popular downtown areas, likely serve a mix of students, commuters, and tourists who may need reliable access to bikes despite adverse weather. Given the high trip counts under these conditions, these stations may benefit from better bike stocking and maintenance during extreme weather events to accommodate the consistent demand and prevent shortages.

**Which stations could benefit from additional bikes during peak hours?**

To identify the busiest bikeshare station during peak hours in Austin, highlighting locations that might benefit from additional bikes to meet high demand. We firstly calculate the number of trips starting during morning (7-9 AM) and evening (4-6 PM) peak hours for each station using a Common Table Expression named `PeakHour`. Then, a second Common Table Expression named `PeakDemand`, aggregates these counts to find the total peak-hour trip count per station, ensuring stations with no peak-hour trips are assigned a count of zero. In the main query, stations with above-average peak-hour demand are filtered and sorted in descending order by trip count, with the `LIMIT 1` clause returning only the single busiest station. (NOTE: why we compare with average demand: By comparing each station's number of trips during peak hours to the average peak-hour demand across all stations, we can focus on stations with higher-than-average demand, which are more likely to experience shortages and need extra bikes.)

```python
%%bigquery
WITH PeakHour AS (
 SELECT
   start_station_id,
   COUNT(trip_id) AS trip_count,
   CASE
     WHEN EXTRACT(HOUR FROM start_time) BETWEEN 7 AND 9 THEN 'Morning Peak'
     WHEN EXTRACT(HOUR FROM start_time) BETWEEN 16 AND 18 THEN 'Evening Peak'
   END AS peak_period
 FROM
   `bigquery-public-data.austin_bikeshare.bikeshare_trips`
 WHERE
   EXTRACT(HOUR FROM start_time) BETWEEN 7 AND 9
   OR EXTRACT(HOUR FROM start_time) BETWEEN 16 AND 18
 GROUP BY
   start_station_id,
   peak_period
),


PeakDemand AS (
 SELECT
   pd.station_id,
   pd.name AS station_name,
   COALESCE(SUM(ph.trip_count), 0) AS peak_trip_count
 FROM
   `bigquery-public-data.austin_bikeshare.bikeshare_stations` AS pd
 LEFT JOIN
   PeakHour AS ph
 ON
   pd.station_id = ph.start_station_id
 GROUP BY
   pd.station_id,
   pd.name
)


SELECT
 station_id,
 station_name,
 peak_trip_count
FROM
 PeakDemand
WHERE
 peak_trip_count > (SELECT AVG(peak_trip_count) FROM PeakDemand)
ORDER BY
 peak_trip_count DESC;
```

The data on peak trip counts across bike stations suggests several high-demand locations that could benefit from additional bikes during peak hours to meet rider needs. The 21st & Speedway @PCL station stands out significantly with a peak trip count of 54,072, indicating a very high volume of usage, likely due to its proximity to high-traffic areas such as university buildings or public facilities, and would benefit greatly from an increase in bike availability to reduce potential shortages. Other notable stations with high peak usage include 4th/Sabine (29,016 trips), Nueces & 26th (22,650 trips), and Rainey St @ Cummings (22,269 trips). These locations likely serve key destinations or transit hubs, suggesting an ongoing, high-demand flow that could be more efficiently met with additional bikes. Stations around the University of Texas, like UT West Mall @ Guadalupe (18,177 trips) and Dean Keeton & Whitis (18,127 trips), along with popular tourist and recreational spots such as Zilker Park (14,999 trips), also show significant demand, especially during peak hours. Enhancing bike supply at these stations could improve availability and customer satisfaction, especially in periods of high ridership.

**What are the least active stations that could be relocated to higher-demand areas?**

We have already identified the stations that need more bike stock, so what could we do to optimize the stock of shared bikes so that there are enough shared bikes in busy areas? One of the effective ways is to relocate the bikes from least active stations to higher-demand stations. To figure out the single least active bikeshare station in Austin with below-average trip demand, we first calculate the total number of trips for each station using the `TotalTrip` Common Table Expression and then joins this data with the `bikeshare_stations` table in the `LeastActive` Common Table Expression, filtering out stations with no recorded trips (the Null value). The main query selects stations with a trip count below the average across all stations, ordering them by the lowest trip count and limiting the result to the top 10 least active stations. The output provides the station ID, name, and total trip count for these least active stations.

```python
%%bigquery
WITH TotalTrip AS (
 SELECT
   start_station_id,
   COUNT(trip_id) AS trip_count
 FROM
   `bigquery-public-data.austin_bikeshare.bikeshare_trips`
 GROUP BY
   start_station_id
),


LeastActive AS (
 SELECT
   station_info.station_id,
   station_info.name AS station_name,
   TotalTrip.trip_count
 FROM
   `bigquery-public-data.austin_bikeshare.bikeshare_stations` AS station_info
 LEFT JOIN
   TotalTrip
 ON
   station_info.station_id = TotalTrip.start_station_id
 WHERE
   TotalTrip.trip_count IS NOT NULL
)


SELECT
 station_id,
 station_name,
 trip_count
FROM
 LeastActive
WHERE
 trip_count < (SELECT AVG(trip_count) FROM TotalTrip)
ORDER BY
 trip_count ASC
LIMIT 10;
```

The analysis identifies the 10 least active bikeshare stations in Austin, with below-average trip counts. For example, the least active bikeshare station in Austin with below-average trip demand is the station with ID 1001, named "OFFICE/Main/Shop/Repair", with a total trip count of only 12. This very low trip count indicates that this station has minimal usage compared to other stations in the network. For the stations that have relatively low utilization, they might be considered for relocation to higher-demand areas.


**Which station needs more bike stock during the weekend?**

This SQL code identifies the top 10 bikeshare stations with the highest demand for bikes on weekends by analyzing trip patterns. It first filters trips starting on weekends (Saturdays and Sundays) and groups them by station ID, using ANY_VALUE to ensure a single station name is associated with each ID while counting the total trips per station (weekend_trip_count). The WeekendDemand step aggregates these counts across all weekends to calculate the total weekend trips (total_weekend_trips) for each station. The HighDemandStations step ranks stations by total weekend trips using the RANK() function, and the final query selects the top 10 busiest stations based on their weekend demand. The results provide station IDs, names, and total weekend trips, helping identify high-demand stations where bike availability needs to be prioritized.


```python
%%bigquery
WITH WeekendTrips AS (
 -- Filter trips that start on weekends and count trips per station
 SELECT
   start_station_id,
   ANY_VALUE(start_station_name) AS start_station_name, -- Avoid duplicates caused by slight naming differences
   COUNT(trip_id) AS weekend_trip_count
 FROM
   `bigquery-public-data.austin_bikeshare.bikeshare_trips`
 WHERE
   EXTRACT(DAYOFWEEK FROM start_time) IN (7, 1) -- 7 = Saturday, 1 = Sunday
 GROUP BY
   start_station_id
),


WeekendDemand AS (
 -- Aggregate trip counts for each station across all weekends
 SELECT
   start_station_id,
   start_station_name,
   SUM(weekend_trip_count) AS total_weekend_trips
 FROM
   WeekendTrips
 GROUP BY
   start_station_id, start_station_name
),


HighDemandStations AS (
 -- Rank stations based on weekend demand
 SELECT
   start_station_id,
   start_station_name,
   total_weekend_trips,
   RANK() OVER (ORDER BY total_weekend_trips DESC) AS rank_by_weekend_demand
 FROM
   WeekendDemand
)


-- Final query: focus on high-demand stations
SELECT
 start_station_id,
 start_station_name,
 total_weekend_trips
FROM
 HighDemandStations
WHERE
 rank_by_weekend_demand <= 10 -- Top 10 stations with the highest weekend demand
ORDER BY
 total_weekend_trips DESC;
```

The output identifies the top 10 bikeshare stations in Austin with the highest demand for bikes on weekends, based on total trips starting from each station. The station "21st/Speedway @ PCL" (ID: 3798) is the busiest, with a total of 42,082 weekend trips, followed by "Riverside/South Lamar" (ID: 2575) with 27,829 trips and "Rainey/Cummings" (ID: 2707) with 24,839 trips. Other notable stations include "Zilker Park" (ID: 2574) with 23,911 trips and "Rainey/Driskill" (ID: 2563) with 19,120 trips. The list indicates that these stations experience consistently high weekend demand, likely due to their proximity to popular areas such as parks, nightlife spots, or university campuses. These insights suggest that these stations should be prioritized for bike stocking and rebalancing during weekends to meet user demand effectively.

```python
%%bigquery
WITH BikeTrips AS (
  SELECT
    DATE(start_time) AS trip_date,
    COUNT(*) AS bike_trip_count
  FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_trips`
  GROUP BY
    trip_date
),
WeatherData AS (
  SELECT
    DATE(datetime) AS weather_date,
    CASE
      WHEN conditions LIKE '%Clear%' THEN 'Sunny'
      ELSE 'Not Sunny'
    END AS weather_condition
  FROM
    `fall24-ba775-a10.ba_775_a10_project.austin_weather`
),
CombinedData AS (
  SELECT
    bt.trip_date,
    bt.bike_trip_count,
    wd.weather_condition
  FROM
    BikeTrips bt
  LEFT OUTER JOIN
    WeatherData wd
  ON
    bt.trip_date = wd.weather_date
)
SELECT
  weather_condition,
  AVG(bike_trip_count) AS avg_bike_trips
FROM
  CombinedData
GROUP BY
  weather_condition;
```

**Could there be any environmental benefits associated with bike share usage trends?**

The goal here was to figure out the environmental benefits of bike share usage by estimating how much CO2 is saved by people choosing bikes over cars. The query was set up to calculate CO2 reduction based on the average bike trip duration, assuming 0.1 kg of CO2 is saved for every kilometer not driven by car. The results show a range of CO2 reductions, with values like 2.87 kg on December 12, 2013, and up to 1,262 kg on June 24, 2024. This shows that on days with more bike trips, there's a noticeable reduction in CO2 emissions, but it also depends on how many people actually use the bikes that day. It's clear that bike shares do have an environmental benefit, but the numbers seem to vary a lot depending on the day.

```python
%%bigquery
WITH BikeTrips AS (
  SELECT
    DATE(start_time) AS trip_date,
    COUNT(*) AS bike_trip_count
  FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_trips`
  GROUP BY
    trip_date
),
AverageTripDistance AS (
  SELECT
    AVG(duration_minutes) as avg_trip_duration
  FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
),
EstimatedEmissions AS (
  -- Estimate CO2 emissions for car trips based on average trip duration.
  SELECT
    b.trip_date,
    b.bike_trip_count * a.avg_trip_duration * 0.1 as estimated_co2_reduction -- Assume 0.1 kg CO2 per km for a car trip
  FROM
    BikeTrips b
  CROSS JOIN
    AverageTripDistance a
)

SELECT
  trip_date,
  estimated_co2_reduction
FROM
  EstimatedEmissions
ORDER BY
  trip_date;
```

**How does the average trip duration change with temperature?**

To determine how the average trip duration changes with temperature, we can join the Bikeshare_trips and weather tables by date, allowing us to analyze bikeshare trip durations in relation to monthly average temperatures.
We select the average temperature and calculate the average trip duration (AVG(bt.duration_minutes)) for each month.
And then grouping by month and average temperature lets us calculate daily averages.

```python
%%bigquery
SELECT
    DATE_TRUNC(w.datetime, MONTH) AS month,
    AVG(w.temp) AS monthly_avg_temperature,
    AVG(bt.duration_minutes) AS monthly_avg_trip_duration
FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS bt
JOIN
    `fall24-ba775-a10.ba_775_a10_project.austin_weather` AS w
ON
    DATE(bt.start_time) = w.datetime
GROUP BY
    month
ORDER BY
    month;
```

**Are trip durations typically longer or shorter on weekends?**

To determine if trip durations are typically longer or shorter on weekends, we could categorize each trip as either a "Weekday" or "Weekend" based on the trip's start date. It then calculates the average trip duration for each category. Specifically, the query extracts the day of the week from the trip's start time, designates trips occurring on Saturday or Sunday as "Weekend," and labels the remaining days as "Weekday." The average trip duration is then computed for both "Weekend" and "Weekday" categories, allowing us to compare and identify if trips taken during weekends generally have longer or shorter durations than those taken on weekdays.


```python
%%bigquery
SELECT
    CASE
        WHEN EXTRACT(DAYOFWEEK FROM bt.start_time) IN (1, 7) THEN 'Weekend'
        ELSE 'Weekday'
    END AS day_type,
    AVG(bt.duration_minutes) AS avg_trip_duration
FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS bt
GROUP BY
    day_type;
```

**Is there a threshold for wind speed above which trip durations decrease?**

To determine if there is a wind speed threshold above which trip durations decrease, we could group data into ranges of wind speeds (e.g., groups we have here are  0-10 km/h, 10-20 km/h, 20-30 km/h, and above 30 km/h) and calculates the average trip duration within each wind speed range. We joined  trip data with weather data based on date, groups the results by wind speed ranges, and then calculates the average trip duration within each group. This approach helps to identify whether there is a specific wind speed threshold where trip durations begin to decrease, providing insights into how wind conditions may influence bikeshare usage patterns.

```python
%%bigquery
SELECT
    CASE
        WHEN w.windspeed <= 10 THEN '0-10 km/h'
        WHEN w.windspeed > 10 AND w.windspeed <= 20 THEN '10-20 km/h'
        WHEN w.windspeed > 20 AND w.windspeed <= 30 THEN '20-30 km/h'
        ELSE 'Above 30 km/h'
    END AS wind_speed_range,
    AVG(bt.duration_minutes) AS avg_trip_duration
FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS bt
JOIN
    `fall24-ba775-a10.ba_775_a10_project.austin_weather` AS w
ON
    DATE(bt.start_time) = w.datetime
GROUP BY
    wind_speed_range
ORDER BY
    wind_speed_range;
```

**What percentage of trips originate from the top 20 busiest stations?**

To calculate the percentage of trips originating from the top 20 busiest stations, we use a Common Table Expression (CTE) to first identify these stations based on the number of rides (num_ride). The bikeshare_stations and bikeshare_trips tables are joined on station_id and start_station_id, and rides are grouped by station_id. The resulting rides are ordered by num_ride DESC to identify the top 20 stations.

Within the main query, the total number of rides from these top 20 stations is summed (SUM(num_ride)), and the total number of trips in the system is obtained using a subquery (COUNT(*)). The ratio of rides from the top 20 stations to the total number of trips is then multiplied by 100 to compute the percentage of trips originating from the busiest stations. This provides insight into the concentration of demand within the system.

```python
%%bigquery
SELECT
    station_id,
    COUNT(*) AS num_ride
FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_stations` AS b
    JOIN
    `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS t
ON
    b.station_id = t.start_station_id
GROUP BY
    station_id
ORDER BY
    num_ride DESC
LIMIT 20;
```

```python
%%bigquery
WITH tmp AS (
    SELECT
        station_id,
        COUNT(*) AS num_ride
    FROM
        `bigquery-public-data.austin_bikeshare.bikeshare_stations` AS b
    JOIN
        `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS t
    ON
        b.station_id = t.start_station_id
    GROUP BY
        station_id
    ORDER BY
        num_ride DESC
    LIMIT 20
)
SELECT
    SUM(num_ride) AS top_20_rides,
    (SUM(num_ride) / (SELECT COUNT(*) FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`)) * 100 AS percentage
FROM
    tmp;
```

**How does bike usage differ across stations between weekdays and weekends?**

To calculate the number of trips originating from each station on weekdays and weekends, the query first joins the bikeshare_stations and bikeshare_trips tables using the station_id and start_station_id fields. This allows for linking station details with their corresponding trip data.

The query then uses COUNTIF to separate trips into two categories based on the day of the week, extracted using EXTRACT(DAYOFWEEK FROM t.start_time). Days 2 (Monday) through 6 (Friday) are counted as weekdays, while Days 1 (Sunday) and 7 (Saturday) are counted as weekends. The results are grouped by station_id and name, representing each station's unique identifier and name. Finally, the query sorts the results by weekday_count in descending order, followed by weekend_count in descending order, to prioritize stations with the highest weekday and weekend trip counts.

```python
%%bigquery
SELECT station_id, name AS station_name,
COUNTIF(EXTRACT(DAYOFWEEK FROM t.start_time) IN (2, 3, 4, 5, 6)) AS weekday_count,
COUNTIF(EXTRACT(DAYOFWEEK FROM t.start_time) IN (1, 7)) AS weekend_count
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations` s
LEFT JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` t
ON s.station_id = t.start_station_id
GROUP BY station_id, name
ORDER BY weekday_count DESC , weekend_count DESC
;
```

The result highlights stations with the most trips on weekdays and weekends, showcasing their demand patterns. **Are there specific stations where trips predominantly occur in one direction?**

To assess the directional flow of trips at each station, the query joins the bikeshare_stations and bikeshare_trips tables on station_id. The join condition accounts for trips either starting (start_station_id) or ending (end_station_id) at a station, with station_id cast as a string to align data types and ensure compatibility.

The query calculates the total trips originating from each station (COUNT(t.start_station_id) AS num_start_trips) and the total trips terminating at each station (COUNT(t.end_station_id) AS num_end_trips). It computes the net directional flow, called trip_balance, as the difference between trips starting and trips ending at a station (num_start_trips - num_end_trips).

The results are grouped by station_id and station_name to uniquely identify each station. Stations are then ranked by their trip balance in descending order, highlighting those with the largest surplus of starting trips compared to ending trips. This method is designed to pinpoint stations exhibiting a strong directional trip bias, providing insights into imbalanced usage patterns.

```python
%%bigquery
WITH StartCounts AS (
    SELECT
        t.start_station_id AS station_id,
        COUNT(*) AS num_start_trips
    FROM
        `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS t
    GROUP BY
        t.start_station_id
),
EndCounts AS (
    SELECT
        t.end_station_id AS station_id,
        COUNT(*) AS num_end_trips
    FROM
        `bigquery-public-data.austin_bikeshare.bikeshare_trips` AS t
    GROUP BY
        t.end_station_id
)
SELECT
    s.station_id,
    s.name AS station_name,
    COALESCE(sc.num_start_trips, 0) AS num_start_trips,
    COALESCE(ec.num_end_trips, 0) AS num_end_trips,
    COALESCE(sc.num_start_trips, 0) - COALESCE(ec.num_end_trips, 0) AS trip_balance
FROM
    `bigquery-public-data.austin_bikeshare.bikeshare_stations` AS s
LEFT JOIN
    StartCounts AS sc
ON
    s.station_id = sc.station_id
LEFT JOIN
    EndCounts AS ec
ON
    CAST(s.station_id AS STRING) = ec.station_id
ORDER BY
    trip_balance DESC;
```

The result highlights stations in the Austin Bike Share system with significant trip imbalances, showcasing directional demand patterns. Stations with a high positive trip_balance (more trips starting than ending) likely serve as major origin hubs, potentially located in residential areas, near universities, or other locations where users initiate trips. Conversely, stations with a high negative balance (more trips ending than starting) could indicate popular destination hubs, such as workplaces, entertainment districts, or transit connections. However, in this dataset, significant negative balances are rare, with most stations exhibiting neutral or positive balances.

Interestingly, a few stations exhibit substantial positive balances, such as 21st & Speedway @PCL, which recorded 3,845 more trips starting than ending. This suggests it is a highly utilized origin station, likely near a residential or student-heavy area. On the other hand, there are no stations with substantial negative trip balances, indicating an overall bias toward starting trips over ending them at specific locations. Stations with near-zero balances, such as Barton Springs & Riverside, indicate relatively balanced use for both trip start and end points, often implying diverse utility across user types and trip purposes.

Many columns had nulls in the weather dataset and initially, we figured this may be a very big issue but some analysis revealed even the nulls have meaning and so, we decided to keep it. The next challenge was to identify how to divide the analysis in different sections such that it gives a holistic and broad view for the stakeholders, who own this bikeshare system and since, they have to deal with city planners and deal with different areas, so how can we cover those aspects and that issue brought the Environmental and Health Insight section. Another big challenge that we encountered was how do we dive deep into the analysis for why students use the system the most, which was uncovered in the Subscriber Type Behavior section. Also, it was a major challenge to identify how to personalise the Weather Specific Trends for Austin but some analysis into the dataset revealed the different aspects which can be used. The windgust, cloudcover and humidity were identified to be the best as Austin is typically known for tornadoes and these are some of the strongest identifiers for the tornadoes.
